```python
from matplotlib import pyplot as plt
import numpy as np

# Generate 100 random data points along 3 dimensions
x, y, scale = np.random.randn(3, 100)
fig, ax = plt.subplots()

# Mapping onto a scattergraph
ax.scatter(x=x, y=y, c=scale, s=np.abs(scale)*500)
ax.set(title="Some Random Data")
plt.show()
```

```python
#Plotting the histograms of the random data
plt.figure()

plt.hist(x, bins=50, alpha=0.4)
plt.hist(y, bins=50, alpha=0.4)
plt.show()
```
